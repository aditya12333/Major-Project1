### Data Description

The columns stands for:

**ID:** *ID of each client*

**MARRIAGE:** *Marital status (1=married, 2=single, 3=others)*

**AGE:** *Age in years*

**PAY_0:** *Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, â€¦             8=payment delay for eight months, 9=payment delay for nine months and above)*

**PAY_2:** *Repayment status in August, 2005 (scale same as above)*

**PAY_3:** *Repayment status in July, 2005 (scale same as above)*

**PAY_4:** *Repayment status in June, 2005 (scale same as above)*

**PAY_5:** *Repayment status in May, 2005 (scale same as above)*

**PAY_6:** *Repayment status in April, 2005 (scale same as above)*

**BILL_AMT1:** *Amount of bill statement in September, 2005 (NT dollar)*

**BILL_AMT2:** *Amount of bill statement in August, 2005 (NT dollar)*

**BILL_AMT3:** *Amount of bill statement in July, 2005 (NT dollar)*

**BILL_AMT4:** *Amount of bill statement in June, 2005 (NT dollar)*

**BILL_AMT5:** *Amount of bill statement in May, 2005 (NT dollar)*

**BILL_AMT6:** *Amount of bill statement in April, 2005 (NT dollar)*

**PAY_AMT1:** *Amount of previous payment in September, 2005 (NT dollar)*

**PAY_AMT2:** *Amount of previous payment in August, 2005 (NT dollar)*

**PAY_AMT3:** *Amount of previous payment in July, 2005 (NT dollar)*

**PAY_AMT4:** *Amount of previous payment in June, 2005 (NT dollar)*

**PAY_AMT5:** *Amount of previous payment in May, 2005 (NT dollar)*

**PAY_AMT6:** *Amount of previous payment in April, 2005 (NT dollar)

**default_payment** (1=yes, 0=no)

# Data Cleaning

### Marriage

There are 3 categories in marriage. 1, 2 & 3 but there is 0 present. Need to impute or delete it

```python
df['MARRIAGE'].value_counts()
```

Only 54 numbers of 0s are there. So those can be removed.

Replacing 0s with null values and then drop them

```python
df.drop(df[df['MARRIAGE'] == 0].index, inplace=True)
```

```python
df['MARRIAGE'].value_counts()
```

'MARRIAGE' column has been cleaned

### Education

In this column has 6 attributes but 5 & 6 means the same. So, it is futile to continue with two different attributes for the same purpose.

```python
df['EDUCATION'].value_counts()
```

Now in this column we have **14** 0s which can be removed. As, there are only **51** 6s, we can just remove it

```python
df.drop(df[df['EDUCATION'] == 0].index, inplace=True)
```

```python
df.drop(df[df['EDUCATION'] == 6].index, inplace=True)
```

```python
df['EDUCATION'].value_counts()
```

'EDUCATION' column is now cleaned

```python
x = df.drop(['Default Payment'], axis = 1)
y = df["Default Payment"]
```

```python
# scaling the feature
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 1)
```

## Oversampling

```python
#!pip install imblearn
from imblearn.over_sampling import RandomOverSampler
oversample = RandomOverSampler(0.75)
x_res, y_res = oversample.fit_resample(x_train, y_train)
print(x_res.shape,y_res.shape)
```

```python
# Count the number of classes 
from collections import Counter
print("The number of classes before fit {}".format(Counter(y)))
print("The number of classes after fit {}".format(Counter(y_res)))
```

```python
plt.figure(figsize = (10,5))
sns.countplot(x = y_res, data=df)
plt.xticks(fontsize=12, rotation=0)
plt.yticks(fontsize=12)
plt.show()
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
Lg = LogisticRegression()
Lg.fit(x_train, y_train)
```

```python
Lg_pred = Lg.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, Lg_pred))
print(metrics.roc_auc_score(y_test,Lg_pred))
print(metrics.confusion_matrix(y_test, Lg_pred))
print(metrics.classification_report(y_test, Lg_pred))
```

## Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(x_res, y_res)
rf_pred = rf.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test, rf_pred))
print(metrics.roc_auc_score(y_test, rf_pred))
print(metrics.confusion_matrix(y_test, rf_pred))
print(metrics.classification_report(y_test, rf_pred))
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_res, y_res)
nb_pred = nb.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test, nb_pred))
print(metrics.roc_auc_score(y_test, nb_pred))
print(metrics.confusion_matrix(y_test, nb_pred))
print(metrics.classification_report(y_test, nb_pred))
```

##  SVM

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_res, y_res)
svc_pred = nb.predict(x_test)

print(metrics.accuracy_score(y_test, svc_pred))
print(metrics.roc_auc_score(y_test, svc_pred))
print(metrics.confusion_matrix(y_test, svc_pred))
print(metrics.classification_report(y_test, svc_pred))
```

## XGB

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(x_res, y_res)
xgb_pred = nb.predict(x_test)

print(metrics.accuracy_score(y_test, xgb_pred))
print(metrics.roc_auc_score(y_test, xgb_pred))
print(metrics.confusion_matrix(y_test, xgb_pred))
print(metrics.classification_report(y_test, xgb_pred))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_res, y_res)
knn_pred = knn.predict(x_test)

print(metrics.accuracy_score(y_test, knn_pred))
print(metrics.roc_auc_score(y_test, knn_pred))
print(metrics.confusion_matrix(y_test, knn_pred))
print(metrics.classification_report(y_test, knn_pred))
```
